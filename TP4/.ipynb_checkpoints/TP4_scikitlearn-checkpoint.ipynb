## Données de travail
**Jeu de données:** "fromage1.txt" comprend 29 instances de fromage,
représentées par 9 attributs <br>
*(calories, sodium, calcium, lipides, retinol, folates, proteines,
cholesterol et magnesium).*

```python
import pandas as pd
fromage = pd.read_table(r"fromage1.txt", sep="\t", header=0, index_col=0) # fromage dataframe
# Dataset
print(fromage)
```

```python
print(f'Dataset shape: {fromage.shape}')
print(f'Dataset description: \n {fromage.describe()}')
```

### Méthode des centres mobiles

```python
import numpy as np
from sklearn import cluster
np.random.seed(0) # la reproduction des mêmes valeurs à chaque itération
# Initialisation
kmeans = cluster.KMeans(n_clusters = 4)
kmeans.fit(fromage)
# Trier selon l'index du label de chaque fromage
ids = np.argsort(kmeans.labels_)
print(pd.DataFrame(fromage.index[ids], kmeans.labels_[ids])) # visualiser la classification
print("Distances \n", kmeans.transform(fromage)) # visialise distances to cluster centers
print("cluster centers \n",kmeans.cluster_centers_)
```

#### Métrique silhouette:

```python
from sklearn import metrics

res = np.arange(9, dtype ="double")
for k in np.arange(9):
  km = cluster.KMeans(n_clusters = k+2)
  km.fit(fromage)
  res[k] = metrics.silhouette_score(fromage,km.labels_)
print (res)
# Display
import matplotlib.pyplot as plt
plt.title("silhouette")
plt.xlabel("# of clusters")
plt.plot(np.arange(2,11,1),(res))
plt.show
```

Le score le plus élévé est présque de 0.5 pour un nombre de clusters de 2.

### Classification ascendante hiérarchique

```python
import pandas
import numpy as np
#librairies pour la CAH
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

Z = linkage(fromage,method='ward', metric='euclidean')# calculer les groupements en clusters du dataset

#affichage du dendrogramme
plt.figure(figsize=(10,10)) # cette ligne permet d'augmenter la résolution du dendogramme pour un affichage plus net
plt.title("CAH")
plt.title('CAH avec matérialisation des 4 classes')# cette ligne ajoute un titre à la figure
dendrogram(Z,labels=fromage.index,orientation='left',color_threshold=255) # créer le dendogramme qui représente les clusters
plt.show()

groupes_cah = fcluster(Z, t = 255,criterion='distance')# diviser le dataset en clusters dont la distance <= seuil 255
print(groupes_cah)# afficher les groupes calculés par la ligne précédente

#index triés des groupes
idg = np.argsort(groupes_cah)
#affichage des observations et leurs groupes
print(pandas.DataFrame(fromage.index[idg],groupes_cah[idg]))
```

**interpretation:** On divise le dataset en 4 clusters avec un seuil 255.

```python
# Correspondance entre les groupes de la CAH et les clusters de Kmeans en
pd.crosstab(groupes_cah, kmeans.labels_, rownames=['CAH'], colnames=['K-means',])
```

Cluster 1 de CAH | Cluster 2 de K-means
Cluster 2 de CAH | Cluster 1 de K-means
Cluster 3 de CAH | Cluster 3 de K-means
Cluster 4 de CAH | Cluster 0 de K-means
